```python
model_name = "/kaggle/input/gemma-2/transformers/gemma-2-2b-it/1"
```

```python
dataset_name = "hydrated_sampled_dataset.jsonl"
dataset_path = f"/kaggle/working/{dataset_name}"
use_hf = False
```

```python
training_config = {
    "model": {
        "pretrained_name": model_name,
        "max_length" : 20000
    },
    "datasets": {
        "use_hf": use_hf,
        "path": dataset_path
    },
    "verbose": True
}
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
import torch
from datasets import Dataset, DatasetDict

# Load the dataset
dfnew = pd.read_json(dataset_path, lines=True)

# Split the dataset into 80% training and 20% testing
train_df, test_df = train_test_split(dfnew, test_size=0.2, random_state=42)

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

# Define a function to tokenize data in batches and ensure all tokens are of length 20000
def tokenize_and_split_data_in_batches(df, tokenizer, fixed_length=20000):
    tokenized_data = {
        'question': [],
        'answer': [],
        'input_ids': [],
        'attention_mask': [],
        'labels': []
    }

    # Process the data
    inputs = df['input'].tolist()
    targets = df['response'].tolist()

    # Ensure both inputs and targets are lists of strings
    inputs = [str(input_text) for input_text in inputs]
    targets = [str(target_text) for target_text in targets]

    # Tokenize the inputs and targets
    tokenized_inputs = tokenizer(inputs, padding='max_length', truncation=True, return_tensors="pt", max_length=fixed_length)
    tokenized_targets = tokenizer(targets, padding='max_length', truncation=True, return_tensors="pt", max_length=fixed_length)

    # Store tokenized data
    tokenized_data['question'].extend(inputs)
    tokenized_data['answer'].extend(targets)
    tokenized_data['input_ids'].extend(tokenized_inputs['input_ids'])
    tokenized_data['attention_mask'].extend(tokenized_inputs['attention_mask'])
    tokenized_data['labels'].extend(tokenized_targets['input_ids'])

    return tokenized_data

# Tokenize the training and testing datasets
train_tokenized = tokenize_and_split_data_in_batches(train_df, tokenizer, fixed_length=20000)
test_tokenized = tokenize_and_split_data_in_batches(test_df, tokenizer, fixed_length=20000)

# Convert the tokenized data into a Dataset object
train_dataset = Dataset.from_dict(train_tokenized)
test_dataset = Dataset.from_dict(test_tokenized)

# Combine the training and testing datasets into a DatasetDict
dataset_dict = DatasetDict({
    'train': train_dataset,
    'test': test_dataset
})

print("Training Dataset:")
print(train_dataset)

print("\nTesting Dataset:")
print(test_dataset)
```

```python
def inference(text, model, tokenizer, max_input_tokens=20000, max_output_tokens=4000):
  # Tokenize
  input_ids = tokenizer.encode(
          text,
          return_tensors="pt",
          truncation=True,
          max_length=max_input_tokens
  )

  # Generate
  device = model.device
  generated_tokens_with_prompt = model.generate(
    input_ids=input_ids.to(device),
    max_length=max_output_tokens
  )

  # Decode
  generated_text_with_prompt = tokenizer.batch_decode(generated_tokens_with_prompt, skip_special_tokens=True)

  # Strip the prompt
  generated_text_answer = generated_text_with_prompt[0][len(text):]

  return generated_text_answer
```

```python
training_args = TrainingArguments(

  # Learning rate
  learning_rate=1.0e-5,

  # Number of training epochs
  num_train_epochs=30,

  # Max steps to train for (each step is a batch of data)
  # Overrides num_train_epochs, if not -1
  max_steps=max_steps,

  # Batch size for training
  per_device_train_batch_size=1,

  # Directory to save model checkpoints
  output_dir=output_dir,

  # Other arguments
  overwrite_output_dir=False, # Overwrite the content of the output directory
  disable_tqdm=False, # Disable progress bars
  eval_steps=120, # Number of update steps between two evaluations
  save_steps=120, # After # steps model is saved
  warmup_steps=1, # Number of warmup steps for learning rate scheduler
  per_device_eval_batch_size=1, # Batch size for evaluation
  evaluation_strategy="steps",
  logging_strategy="steps",
  logging_steps=1,
  optim="adafactor",
  gradient_accumulation_steps = 4,
  gradient_checkpointing=False,

  # Parameters for early stopping
  load_best_model_at_end=True,
  save_total_limit=1,
  metric_for_best_model="eval_loss",
  greater_is_better=False
)
```
